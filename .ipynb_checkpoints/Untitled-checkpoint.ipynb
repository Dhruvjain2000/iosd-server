```python
import requests
from bs4 import BeautifulSoup as BS
import json

url = 'http://localhost:8080/Home.html'
r = requests.get(url)
```

```python
soup = BS( r.text  , 'html5lib' )
```

```python
print(soup.prettify())
```

```python
items  = soup.findAll('div' , { 'class' : 'vid-item'})
```

```python
item = items[0]
```

```python
def extract(item) :
    di = {}
    di['link'] = item['onclick'].split('=')[1].replace("'" , '')
    di['image'] = item.findAll( 'div' , {'class' : 'thumb' } )[0].img['src'].replace('Home_files' , 'thumbnails')
    di['title'] = item.findAll( 'div' , {'class' : 'desc' } )[0].text
    
    return di
    
    
    
```

```python
extract(item)
```

```python
data = [ extract(item) for item in items ]
```

```python
data
```

```python
fs = open('data.json' , 'w')
```

```python
fs.write( json.dumps( data  , indent=4) )
```

```python
fs.close()
```
